### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
AGE = HR['Age']
QUARTILE_75 = AGE.quantile(q=0.75)
QUARTILE_25 = AGE.quantile(q=0.25)
IQR = QUARTILE_75 - QUARTILE_25
# print("25th Quartile = ", QUARTILE_25)
# print("75th Quartile = ", QUARTILE_75)
print("IQR = ", IQR)
```

### Find out the mean, median and variance for the column - DailyRate 

```python
DAILY_RATE = HR['DailyRate']

print("Mean = ", DAILY_RATE.mean())
print("Median = ", DAILY_RATE.median())
print("Variance = ", DAILY_RATE.var())
print("SD =", DAILY_RATE.std())
```
